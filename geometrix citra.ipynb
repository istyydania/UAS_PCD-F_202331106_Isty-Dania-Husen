```python
#202331106_Istydaniahusen
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#202331106_Istydaniahusen
image = cv2.imread('myphoto.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 1. Rotated 90 derajat
rotated = cv2.rotate(image_rgb, cv2.ROTATE_90_CLOCKWISE)

# 2. Resized ke 300x300 px
resized = cv2.resize(image_rgb, (100, 300))

# 3. Cropped
h, w = image_rgb.shape[:2]
crop = image_rgb[h//4:h*3//4, w//4:w*3//4]

# 4. Flipped
flipped = cv2.flip(image_rgb, 1)

# 5. Translated
M = np.float32([[1, 0, 100], [0, 1, 50]])
translated = cv2.warpAffine(image_rgb, M, (w, h))

titles = ['Citra Asli', 'After Rotated', 'After Resized', 'After Cropped', 'After Flipped', 'After Translated']
images = [image_rgb, rotated, resized, crop, flipped, translated]

plt.figure(figsize=(15, 10))
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i])
    plt.title(titles[i], fontsize=12)
    plt.xlabel('Pixel X')
    plt.ylabel('Pixel Y')
    plt.xticks(np.linspace(0, images[i].shape[1], num=5, dtype=int))
    plt.yticks(np.linspace(0, images[i].shape[0], num=5, dtype=int))
    plt.grid(True)
plt.tight_layout()
plt.show()
```
